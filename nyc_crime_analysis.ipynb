```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
from pathlib import Path
import numpy as np
import scipy.stats as st
```

```python
#import data files
nyc_crimedata_path = "clean_data/2021-2022_nyc_crime_data.csv"
nyc_poppath_path = "clean_data/nyc_pop_data.csv"

```

```python
#Read nyc crime data 
nyc_crime_df = pd.read_csv(nyc_crimedata_path)
nyc_crime_df.head()
```

```python
#Read nyc pop data 
nyc_pop_df = pd.read_csv(nyc_poppath_path)
nyc_pop_df.head()
```

```python
#Combine data into single dataframe 
combined_data = pd.merge(nyc_crime_df, nyc_pop_df, on="borough")

combined_data.head()
```

```python
#Rename col headers 

renamed_df = combined_data.rename(columns={"offense_lvl": "offence level", "suspect_age_gp": "suspect age group",
                                            "victim_age_gp": "victim age group", "pop": "borough population"})

renamed_df.head()
```

```python
#age group victim

#convert col to string 
renamed_df['suspect age group'] = renamed_df['suspect age group'].astype(str)
renamed_df['victim age group'] = renamed_df['victim age group'].astype(str)

#group by victim
age_victim = renamed_df.groupby(["victim age group"])["offence level"].count()

age_victim
```

```python
#check col
print("\nTypes of data after conversion:\n", renamed_df.dtypes)
```

```python
#group by suspect
age_suspect = renamed_df.groupby(["suspect age group"])["offence level"].count()

age_suspect
```
